**실습명: NLTK 패키지를 이용한 규칙 기반 구구조 구문 분석**

```python
#NLTK 패키지 다운로드
!pip install nltk==3.3
import nltk
```

```python
#구구조 구문 분석 규칙 직접 작성
grammer = nltk.CFG.fromstring("""
S -> NP VP
NP -> NN XSN JK | NN JK
VP -> NP VP | VV EP EF
NN -> '아이' | '케이크'
XSN -> '들'
JK -> '이' | '를'
VV -> '먹'
EP -> '었'
EF -> '다'
""")
```

```python
#규칙 기반 구문 분석기 생성
#ChartParser(구문 분석 알고리즘) 외에 ShiftReduceParser, RecursiveDescentParser 제공
parser = nltk.ChartParser(grammer)

#구구조 구문 분석 수행
sent = ['아이', '들', '이', '케이크', '를', '먹', '었', '다']
for tree in parser.parse(sent):
  print(tree)
```

**실습명: Spacy를 이용한 의존 구문 분석**

```python
#Spacy 패키지 다운로드
!pip install spacy==2.2.4
import spacy
```

영어 문장 의존 구문 분석<br>
- Spacy 모델은 문장을 token들로 구성된 document로 처리
- 각 token에는 품사, 의존 관계, 개체명 정보 태깅
- 지배소: 의존 관계에서 절의 중심, 의존소: 절 내 지배소에 의존하는 단어

출력정보 종류<br>
- `token.text` = token 문자열
- `token.dep_` = token과 token의 지배소 간의 의존 관계 유형
- `token.head` = 지배소 token

```python
#영어 multi task 통계 모델
nlp = spacy.load('en_core_web_sm') #영어 통계 모델 불러오기
doc = nlp('The fat cat sat on the mat')
for token in doc:
  print(token.text, token.dep_, token.head.text)
```

```python
#의존 분석 결과 시각화
from spacy import displacy

displacy.render(doc, style='dep', jupyter=True)
```
