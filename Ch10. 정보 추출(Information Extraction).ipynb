```python
#구조화된 데이터로 표현
#(엔티티1, 관계, 엔티티2) 형식으로 표현
locs = [('고려대학교', 'In', '서울'),
        ('Naver', 'In', '성남'),
        ('KT 위즈', 'In', '수원'),
        ('삼성', 'In', '서울')]
```

```python
#서울에 있는 기관만 출력
query = []
for (e1, re1, e2) in locs:
  if e2 == '서울':
    query.append(e1)

print(query)
```

# 실습 [10-2]<br>
**실습명: 정규식을 적용한 관계 추출 실습1**<br>
: (엔티티1, 관계, 엔티티2) 형식의 관계형 튜플 표현에서 관계는 엔터티1과 2 사이에 있는 단어의 문자열이다, 이를 기반으로 정규표현식을 사용해 관계의 인스턴스 추출 가능

```python
#관계 추출(relation extraction) 라이브러리 (ieer corpus 사용)
import re #정규표현식 라이브러리
import nltk
nltk.download('ieer')
```

```python
#단어가 포함된 문자열 검색시 사용되는 관계 추출 정규식 = (?!\b.+ing\b)
IN = re.compile(r'.*\bin\b(?!\b.+ing)')

#ORG(기관이름), LOC(장소)를 관계형 튜플 방식으로 나타내기
for doc in nltk.corpus.ieer.parsed_docs('NYT_19980315'):
  print("doc : ", doc)
  for rel in nltk.sem.extract_rels('ORG', 'LOC', doc, corpus='ieer', pattern=IN):
    print(nltk.sem.rtuple(rel))
```

# 실습 [10-3]<br>
**실습명: 정규식을 적용한 관계 추출 실습2**<br>
- conll2002 Dutch 코퍼스에는 명명된 엔티티 주석, 품사 태그 포함
- `cluase()` : 이진 관계 기호가 매개 변수 relsym 값으로 지정되는 관계를 clausal 형식으로 인쇄

```python
#라이브러리
import re
import nltk
from nltk.corpus import conll2002
nltk.download('conll2002')
```

```python
#여러 개의 정규식 추가
vnv = """
(
  is/V|
  was/V|
  werd/V|
  wordt/V
)
.*
van/Prep
"""
```

```python
#정규식 기준으로 전처리
VAN = re.compile(vnv, re.VERBOSE)
```

```python
#conll2002 코퍼스에 대한 관계 추출 (네덜란드어)
for doc in conll2002.chunked_sents('ned.train'):
  #print("doc : ", doc)
  for rel in nltk.sem.extract_rels('PER', 'ORG', doc, 
                                   corpus='conll2002', pattern=VAN):
    print("", nltk.sem.clause(rel, relsym="VAN"))
    #print(nltk.rtuple(rel, lcon=True, rcon=True))
```
