<br>
NLTK(Natural Language Toolkit)를 이용한 개체명 정보 추출:
1. 입력받은 문장을 토큰 단위로 분리
2. 분리된 단어를 형태소 분석
3. 형태소로 분리된 단어 개체명으로 분석

```python
import nltk

nltk.download('punkt')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
```

```python
#1. 입력문장 토큰 단위로 분리
sentence = "Prime Minister Boris Johnson had previously said the UK would leave by 31 October."

tokens = nltk.word_tokenize(sentence)
print(tokens)
```

```python
#2. 분리된 단어 형태소 분석 
#('단어','형태소') 형태로 출력
tagged = nltk.pos_tag(tokens)
print(tagged)
```

```python
#3. 개체명 인식(출력)
entities = nltk.chunk.ne_chunk(tagged)
print(entities)

"""
PERSON: 사람
(Boris와 Johnson이 하나의 개체명으로 인식된 이유: BIO Tagging(Begin,Inside,Outside))
ORGANIZATION: 조직
"""
```
